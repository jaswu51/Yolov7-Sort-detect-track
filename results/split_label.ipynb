```python
import numpy as np
import pandas as pd
```

```python
# read text file into pandas DataFrame
df= pd.read_csv("AvenueDatasetResults.txt", sep=",")
```

```python
df_testing=df.iloc[:141865,:]
df_training= df.iloc[141866:,:]
```

```python
import scipy.io as scio
import numpy as np
```

# get the anormaly labels

```python
df_testing.head(2)
```

```python
video_no=[]
frame_no=[]
sum_pixel_abnormal_value=[]
for i in range(0,21):
    dataFile ="/Users/xiaokeai/Downloads/yolov7-main/datasets/AvenueDataset/evaluation_code/testing_label_mask/" +str(i+1)+ "_label.mat"
    data = scio.loadmat(dataFile)
    for j in range(1,len(data['volLabel'][0])+1):
        video_no.append(i)
        frame_no.append(j)
        sum_pixel_abnormal_value.append(np.sum((data['volLabel'][0][j-1])))
```

```python
df1=pd.DataFrame(video_no, columns=['video_no'])
df2=pd.DataFrame(frame_no, columns=['frame_no'])
df3=pd.DataFrame(sum_pixel_abnormal_value, columns=['sum_pixel_abnormal_value'])
df_testing_anormaly=pd.concat([df1,df2,df3],axis=1)
```

```python
df_testing_anormaly
```

```python
df_testing_index=pd.concat([df_testing['path'],df_testing['video_no'],df_testing['frame_no']],axis=1)
df_testing_index=df_testing_index.drop_duplicates()
df_testing_index
```

```python
df_testing_index['sum_pixel_abnormal_value']=np.array(df_testing_anormaly['sum_pixel_abnormal_value'])
```

```python
df_testing_index1 = pd.DataFrame(None, index=np.arange(df_testing.shape[0]), columns=df_testing_index.columns)
```

```python
resdict={}
list(df_testing_index1.index)
resdict = {list(df_testing_index1.index)[i]: list(range(0, 15324))[i] for i in range(15324)}
```

```python
for i in list(df_testing_index.index):
    df_testing_index1.iloc[i,:]=df_testing_index.iloc[resdict[i],:]
```

```python
df_testing_index1
```

```python
df_testing_index1=df_testing_index1.fillna(method='ffill')
df_testing_index1['abnormal']=np.where(df_testing_index1['sum_pixel_abnormal_value'] > 0, 1,0)
```

```python
df_testing_index1.to_csv(r'AvenueDatasetResultsTesting_Label.txt', header=True, index=False, sep=',', mode='a')
```
